```python
Q1 = np.percentile(data["Salary"], 25)
Q3 = np.percentile(data["Salary"], 75)
    
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
    
outliers = data[(data["Salary"] < lower_bound) | (data["Salary"] > upper_bound)]
```
